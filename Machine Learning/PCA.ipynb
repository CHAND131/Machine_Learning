```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = load_breast_cancer(as_frame=True)
# cancer = df.frame
x = pd.DataFrame(df.data,columns=df.feature_names)
x
```

```python
#standarize
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
scaled_data = stdsc.fit_transform(x)
display(scaled_data)
```

```python
# priciple component analysis
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
p12 = pca.fit_transform(scaled_data)
print(p12.shape)
p12
```

```python
df_pca = pd.DataFrame(p12,columns=["pc1","pc2"])
df_pca
```

```python
plt.scatter(df_pca["pc1"],df_pca["pc2"],c=df.target)
plt.xlabel('First Principal Component - PC1')
plt.ylabel('Second Principal Component - PC2');
```

```python
a = df_pca
b = df.target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
a_train,a_test,b_train,b_test = train_test_split(a,b,test_size=0.2,random_state=0)
```

```python
classifier = LogisticRegression()
classifier.fit(a_train,b_train)
predcit = classifier.predict(a_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, r2_score
print(confusion_matrix(b_test, predcit))
print(classification_report(b_test,predcit))
```
