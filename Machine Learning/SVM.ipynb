```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
iris = load_iris()
# display(iris)
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['target'] = iris.target
display(df)
```

```python
iris.target_names
```

```python
df["flowers_names"] =df.target.apply(lambda  x : iris.target_names[x])
df
```

```python
x=df.drop(["target","flowers_names"],axis=1)
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
iris_svm_model = SVC(kernel="linear",random_state=20)
iris_svm_model.fit(X_train,y_train)
```

```python
y_pred= iris_svm_model.predict(X_test)
iris_svm_model.score(X_test, y_test)
iris_svm_model.score(X_train, y_train)
```
