```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,8,9,6])
y=np.array([2,4,1,5,7,1])
plt.plot(x,y,color='y',linestyle='dotted',marker='*',ms=20,mec='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('My First Graph')
```

```python
plt.bar(x,y,color='y',linestyle='dotted')
```

```python
x = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
y = [1, 4, 3, 2, 7, 6, 9, 8, 10, 5]

plt.plot(x, y, '-r')
plt.xlabel('x')
plt.ylabel('y')
plt.tick_params(axis='x',labelrotation=60)
```

```python
y1 = [2, 3, 4.5]

# Y-axis values
y2 = [1, 1.5, 5]

# Function to plot
plt.plot(y1)
plt.plot(y2)
plt.legend(['green','yellow'],loc='lower right')
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 2, 1)
plt.plot(x,y)

x = np.arange(5)
y = np.arange(5)
plt.subplot(2, 2, 2)
plt.plot(x,y)

x = np.arange(5,10)
y = np.arange(5)
plt.subplot(2, 2, 3)
plt.plot(x,y)

x = np.arange(10,15)
y = np.arange(1,10,2)
plt.subplot(2, 2, 4)
plt.plot(x,y)
```

```python
data = np.random.normal(170, 10, 250)
plt.hist(data,bins=10)
```

```python
z = np.linspace(0, 1, 1000)
x = z * np.sin(800 * z)
y = z * np.cos(800 * z)
ax = plt.axes(projection='3d')

ax.plot3D(x,y,z,'green')
ax.set_title("3D Graph")
```
